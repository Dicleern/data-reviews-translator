## Review Translator

üëâ `reviews` dataset‚Äôini load edin ve 1-yƒ±ldƒ±z rating‚Äôe sahip review‚Äôlardan bir √∂rnek (sample) se√ßin.

```python
# Verileri y√ºkle
from olist.data import Olist
data = Olist().get_data()
```

```python
data.keys()
```

üëÄ 20 adet d√º≈ü√ºk puan alan yorumdan olu≈üan bir √∂rneklem se√ßin (rastgele) ve bunu bir listeye d√∂n√º≈üt√ºr√ºn.

```python
# SENƒ∞N KODUN BURAYA
reviews_df=data['order_reviews']
reviews_df=reviews_df.sort_values(by='review_score', ascending=True)
reviews_df=reviews_df.head(50)
```

```python
reviews_df
```

üó£ Bu g√∂reve ba≈ülamadan √∂nce √∂nceden y√ºklediƒüiniz `google_translator` paketini kullanarak bu yorumlarƒ± √ßevirin:

```python
# SENƒ∞N KODUN BURAYA
from googletrans import Translator
translator = Translator()

# 1. reviews_df i√ßindeki mesajlardan bo≈ü olmayanlarƒ± se√ß ve listeye √ßevir
# head(50) i√ßinden sadece mesajlarƒ± alƒ±yoruz
all_messages = reviews_df['review_comment_message'].dropna().tolist()

# 2. Eƒüer listede 20'den fazla varsa, ilk 20 tanesini alalƒ±m
final_list_to_translate = all_messages[:20]

translated_list = []

print(f"Toplam {len(final_list_to_translate)} ge√ßerli yorum √ßevriliyor...")

# 3. √áeviri D√∂ng√ºs√º
for text in final_list_to_translate:
    try:
        # dest='en' ile ƒ∞ngilizceye √ßevir
        translation = translator.translate(text, dest='en')
        translated_list.append(translation.text)
    except Exception as e:
        translated_list.append("√áeviri Hatasƒ±")

# Sonu√ßlarƒ± g√∂ster
translated_list
```
